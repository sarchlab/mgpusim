```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as sm
```

```python
data = pd.read_csv('r9nano.csv', index_col=0);data
```

```python
grouped = data.groupby('arg')
```

```python
grouped.describe()
```

```python
agg_data = grouped.agg(np.mean)
agg_data = agg_data.reset_index()
agg_data.columns = ['size', 'time']
agg_data
```

```python
result = sm.ols(formula="time ~ size", data=agg_data).fit()
result.params['size']
```
